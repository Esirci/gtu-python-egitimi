# **Adım 1: Ölçümler**

İlk yapacağımız şey sıcaklık ve fan devir hızını ölçüp bir yere kaydetmek. Her bir sıcaklık değerine karşılık gelen aşağıdaki devir hızlarını elde etmiş olalım:

*   30 > 1500
*   33.5 > 1556
*   36.2 > 1692
*   36.3 > 1697
*   40.4 > 1770
*   50.7 > 2241
*   62.4 > 2512
*   75.6 > 2851

Sayılar tamamen hayal ürünüdür.

# **Adım 2: Basit Bir Model Oluşturalım**

```
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib
import matplotlib.pyplot as plt

matplotlib.rcParams["figure.dpi"] = 100
```

```
T = np.array([30.0, 33.5, 36.2, 36.3, 40.4, 50.7, 62.4, 75.6], dtype=float).reshape(-1,1)
S = np.array([1500.0, 1556.0, 1692.0, 1697.0, 1770.0, 2241.0, 2515.0, 2851.0], dtype=float).reshape(-1,1)
```

```
plt.scatter(T, S)
```

```
model = LinearRegression()

model.fit(T, S)
```

```
# Bu kod parçası, np.array fonksiyonu kullanılarak test adında bir NumPy dizisi oluşturur.
# reshape(-1,1): Bu satır, reshape yöntemini kullanarak tek boyutlu diziyi iki boyutlu bir diziye dönüştürür.
# reshape'in ilk parametresi olan -1, NumPy'ın eleman sayısına ve belirtilen sütun sayısına göre satır sayısını otomatik olarak belirlemesini söyler.
# İkinci parametre olan 1, sonuçtaki dizinin bir sütuna sahip olması gerektiğini belirtir.

test = np.array([55.2, 41.5, 70.0, 57.3]).reshape(-1,1)
```

#**Adım 3: Eğitilmiş modeli test edelim**

```
predicted = model.predict(test)
print(predicted)
```

```
plt.scatter(T, S, c='coral')
plt.scatter(test, predicted, c='lightblue')

plt.title('Sıcaklık vs. Devir sayısı')
plt.xlabel('Sıcaklık')
plt.ylabel('Fan devir hızı')
plt.show()
```
