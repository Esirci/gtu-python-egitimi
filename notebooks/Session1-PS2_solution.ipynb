### Geçen derste yaptığımız gibi basit bir hesap makinesi uygulaması oluşturmanız bekleniyor. Bu uygulama; toplama, çıkarma, çarpma ve bölme ve üst alma işlemlerini yapabilmektedir.

### Kullanıcı, bir işlem seçtikten sonra uygulamadan çıkmak isteyene kadar istediği kadar işlem yapabilmelidir. Her işlem seçiminden sonra kullanıcıdan gerekli sayıları alarak işlem sonucunu ekrana yazdırın. Uygulama, döngüler, fonksiyonlar(toplama,çıkarma,vs.) ve kullanıcıdan veri alma yöntemlerini içermelidir.



````python
# Kullanıcıya işlem seçeneklerini gösteren fonksiyon
def show_options():
    print("İşlem Seçenekleri:")
    print("1. Toplama")
    print("2. Çıkarma")
    print("3. Çarpma")
    print("4. Bölme")
    print("5. Üst Alma")
    print("6. Çıkış")

# Toplama işlemini gerçekleştiren fonksiyon
def addition(num1, num2):
    return num1 + num2

# Çıkarma işlemini gerçekleştiren fonksiyon
def subtraction(num1, num2):
    return num1 - num2

# Çarpma işlemini gerçekleştiren fonksiyon
def multiplication(num1, num2):
    return num1 * num2

# Bölme işlemini gerçekleştiren fonksiyon
def division(num1, num2):
    if num2 == 0:
        return "Bölen sıfır olamaz!"
    else:
        return num1 / num2

# Üst alma işlemini gerçekleştiren fonksiyon
def power(base, exponent):
    return base ** exponent
````

````python
#main fonksiyon
def main():
  print("Hesap Makinesi Uygulamasına Hoş Geldiniz!")

  while True:
    show_options()  # Kullanıcıya işlem seçeneklerini göster
    choice = input("Lütfen bir işlem seçin (1/2/3/4/5/6): ")  # Kullanıcıdan işlem seçimini al

    if choice == '6':  # Eğer kullanıcı çıkışı seçtiyse
        print("Hesap makinesi kapatılıyor...")
        break

    num1 = int(input("Birinci sayıyı girin: "))  # İlk sayıyı kullanıcıdan al
    num2 = int(input("İkinci sayıyı girin: "))  # İkinci sayıyı kullanıcıdan al

    # Kullanıcının seçimine göre ilgili işlemi yap ve sonucu ekrana yazdır
    if choice == '1':
        print("Sonuç Toplama:", addition(num1, num2))
    elif choice == '2':
        print("Sonuç Çıkarma:", subtraction(num1, num2))
    elif choice == '3':
        print("Sonuç Çarpma:", multiplication(num1, num2))
    elif choice == '4':
        print("Sonuç Bölme:", division(num1, num2))
    elif choice == '5':
        print("Sonuç Üst Alma:", power(num1, num2))
    else:
        print("Geçersiz giriş! Lütfen 1, 2, 3, 4,5 veya 6'yı seçin.")
````

````python
#main fonksiyonunu kullanarak hesap mak. uygulamasını başlatalım.
main()
````

## Öğrenci Notları ve Harf Notu İstatistikleri Bulma

### Bir öğrenci grubunun notları verildiğinde, her bir harf notunda kaç öğrencinin bulunduğunu ve her bir öğrencinin aldığı notun karşılığı olan harf notlarını hesaplayan bir Python programı yazın.

### Programınız, öğrencilerin adı-soyadı, numarası ve notunu içeren bir sözlük üzerinde gezinmelidir. Program, öğrenci sayısını ve notlarını bir sözlükte depolayacak ve ardından her bir harf notunda kaç öğrencinin olduğunu ekrana yazdıracak.



```
A harf notu alan toplam 4 öğrenci var.
B harf notu alan toplam 2 öğrenci var.
C harf notu alan toplam 3 öğrenci var.
D harf notu alan toplam 3 öğrenci var.
```



````python
# Öğrenci bilgilerini içeren bir sözlük
ogrenciler = {
    "Ali Veli": {"numara": 101, "not": 85},
    "Ayşe Fatma": {"numara": 102, "not": 55},
    "Ahmet Mehmet": {"numara": 103, "not": 75},
    "Zeynep Derya": {"numara": 104, "not": 10},
    "Fatih Selim": {"numara": 105, "not": 65},
    "Mehmet Ali": {"numara": 106, "not": 18},
    "Nur Ayşe": {"numara": 107, "not": 78},
    "Aliye Zeynep": {"numara": 108, "not": 13},
    "Hasan Hüseyin": {"numara": 109, "not": 70},
    "Emre Can": {"numara": 110, "not": 22},
    "Ahmet Kamil": {"numara": 111, "not": 96},
    "Gizem Nur": {"numara": 112, "not": 72},
    "Deniz Burak": {"numara": 113, "not": 14}
}

#Harf notları araligi: A:85-100  B:70-85 C:55-70 D:40-55 F:0-40 (grade_listte tutalım)
# ögrencinin notuna bakarak hangi harf notunu aldigina kara veren fonksiyon yazin.
def harf_notu(notu, grade_list):
    if notu >= grade_list[0]:
        return "A"
    elif notu >= grade_list[1]:
        return "B"
    elif notu >= grade_list[2]:
        return "C"
    elif notu >= grade_list[3]:
        return "D"
    else:
        return "F"

#yeni bir sözlük tanımlayalım
harf_notlari = {}

#harf notlarının alt sınırlarını verdik. A = 85 alt sınır
#grade_list = [85,70,55,40]

#grade_list elemanları kullanıcıdan alınabilir.
#Böylece harf not aralıklarını istediğimiz gibi değiştirebiliriz,
#kodun içini ellemeden farklı sonuçları görebiliriz.

liste = input("Harf notlarının alt sınırarını girin (elemanlar arasında boşluk bırakarak): ")
# Kullanıcının girdisini boşluklara göre ayırarak liste oluştur
grade_list_ch = liste.split()
# Liste elemanlarını tam sayıya dönüştür
grade_list = [];
for eg in grade_list_ch:
  grade_list.append(int(eg))

# Tüm öğrencilerin notlarını kontrol et
for ogrenci in ogrenciler.values():
    # Her öğrencinin notunu al
    ogrenci_notu = ogrenci["not"]

    # Notu harf notuna dönüştür
    harf = harf_notu(ogrenci_notu,grade_list)

    # Eğer harf, harf_notlari sözlüğünde varsa, öğrenci sayısını bir artır
    if harf in harf_notlari:
        harf_notlari[harf]["ogrenci_sayisi"] = harf_notlari[harf]["ogrenci_sayisi"] +1

    # Eğer harf yoksa, yeni bir girdi oluştur ve öğrenci sayısını 1 olarak ayarla
    else:
        harf_notlari[harf] = {"ogrenci_sayisi": 1}

#harf_notları = {A:3,B:4,c:2,D:3,F:1}
# Harf notlarına göre öğrenci sayılarını ve not aralıklarını ekrana yazdır
for harf, bilgiler in harf_notlari.items():
    print(f"{harf} harf notu alan toplam {bilgiler['ogrenci_sayisi']} öğrenci var. ")
````

````python
# Bu kod, bir sözlükteki öğrencilerin notlarını değerlendirir
# belirli bir not aralığında kaç öğrencinin bulunduğunu hesaplar.

def not_araliginda_kac_kisi_var(notlar, alt_sinir, ust_sinir):
    # Belirtilen not aralığındaki öğrenci sayısını hesapla
    sayac = 0
    for not_herbir in notlar:
        if alt_sinir <= not_herbir < ust_sinir:
            sayac += 1
    return sayac


def not_araligi_kontrol(not_herbir, alt_sinir, ust_sinir):
  # Belirtilen not aralığında öğrenci var mı?
  if alt_sinir <= not_herbir < ust_sinir:
      return 1
  else:
    return 0

#kullanıcıdan hangi not aralığına bakmak istedigini alalım.
alt_not = float(input("Alt sınırı girin: "))
ust_not = float(input("Üst sınırı girin: "))

# Başlangıçta, kişi sayısını sıfıra ayarla
kisi_sayisi = 0

# Tüm öğrencilerin notlarını kontrol et
for ogrenci in ogrenciler.values():
    # Her öğrencinin notunu belirli bir aralıkta kontrol et ve bu aralıkta ise kişi sayısını artır
    kisi_sayisi += not_araligi_kontrol(ogrenci["not"], alt_not, ust_not)

# belirtilen not aralığında bulunan kişi sayısını yazdır
print(f"{alt_not} ve {ust_not} arasındaki kisi_sayisi: {kisi_sayisi}")
````
