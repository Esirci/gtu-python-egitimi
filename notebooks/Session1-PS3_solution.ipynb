```python
import numpy as np
import matplotlib.pyplot as plt

# Veri dosyasından öğrenci bilgilerini okuma
with open("ogrenci_notlari.txt", "r") as file:
    lines = file.readlines()


```

```python
# Öğrenci bilgilerini saklamak için boş listeler oluştur
ogrenci_adlari = []
ogrenci_soyadlari = []
vize_notlari = []
final_notlari = []


# Her bir satırdaki verileri ayırma ve saklama
for line in lines:
    #ad, soyad,vize_notu, final_notu = line.strip().split(",")

    ogrenci_adi, vize_notu, final_notu = line.strip().split(",")
    ogrenci_adlari.append(ogrenci_adi)
    vize_notlari.append(float(vize_notu))
    final_notlari.append(float(final_notu))
```

```python

# Numpy dizilerine dönüştürme
#vize_notlari2 = np.array(vize_notlari)
#final_notlari2 = np.array(final_notlari)

# Ortalama ve standart sapma hesaplama
vize_ortalama = np.mean(vize_notlari)
final_ortalama = np.mean(final_notlari)
vize_std = np.std(vize_notlari)
final_std = np.std(final_notlari)

# Histogramlar
plt.figure(figsize=(10, 6))

# Vize sınav notları histogramı
plt.subplot(1, 2, 1)
plt.hist(vize_notlari, bins=11, color='skyblue', edgecolor='black', alpha=0.7)
plt.title('Vize Sınavı Notları')
plt.xlabel('Not Aralığı')
plt.ylabel('Frekans')

# Final sınav notları histogramı
plt.subplot(1, 2, 2)
plt.hist(final_notlari, bins=11, color='lightgreen', edgecolor='black', alpha=0.7)
plt.title('Final Sınavı Notları')
plt.xlabel('Not Aralığı')
plt.ylabel('Frekans')


# Ortalama ve standart sapma grafiği
plt.figure(figsize=(10, 6))

# Vize sınavı notları
plt.bar(['Vize', 'Final'], [vize_ortalama, final_ortalama], yerr=[vize_std, final_std], capsize=10)
plt.title('Ortalama ve Standart Sapma')
plt.xlabel('Sınav Türü')
plt.ylabel('Not Ortalaması')
```

**trend Analiziz & Satış Tahmini**

## Bir perakende şirketi, son birkaç yıldaki aylık satış verilerini içeren bir zaman serisi verisine sahiptir. Şirket, satışların genel bir artış eğilimi olduğunu fark etmiştir, ancak veri setinde bazı aykırı değerler ve gürültüler bulunmaktadır. Şirket, bu veri setindeki lineer bir trendi belirlemek ve gelecekteki satışları tahmin etmek istiyor.

## Bu senaryoda, Python kullanarak bu sorunu nasıl çözümleyeceğinizi açıklayınız. Adım adım düşünerek algortimayı tasarlayınız.

**Kullanabileceğiniz yapılar;**

polyfit ve polyval, Python'un NumPy kütüphanesi içinde bulunan çoklu dereceden polinom regresyonu yapmak için kullanılan fonksiyonlardır.

**polyfit:** Bu fonksiyon, bir veri setindeki x ve y değerlerine en uygun polinom katsayılarını hesaplar. Genellikle numpy.polyfit(x, y, derece) şeklinde kullanılır. x ve y, veri setindeki bağımsız ve bağımlı değişkenleri temsil eder. derece, kullanılacak polinomun derecesini belirtir.

**polyval:** Bu fonksiyon, verilen x değerleri için polinomun y değerlerini hesaplar. numpy.polyval(katsayılar, x) şeklinde kullanılır. katsayılar, polyfit tarafından hesaplanan polinom katsayılarını, x ise tahmin edilecek x değerlerini temsil eder.

```python
#Bu örnek, bir veri setindeki x, y değerlerine en uygun 2. dereceden
#bir polinomun katsayılarını hesaplar ve bu polinomun y değerlerini
#verilen x değerleri için tahmin eder.

import numpy as np

# Veri seti 9x^2 + 6x^1 + 1x^0
#x_ust = np.array([0, 1, 2])
#y = np.array([1, 16, 49])
x_ust = np.array([0, 1, 2, 3, 4, 5])
y = np.array([0, 0.8, 0.9, 0.1, -0.8, -1])

# Polinom katsayılarını hesapla (2. dereceden polinom)
#np.polyfit(x, y, 2) fonksiyonu, verilen x ve y değerlerine en uygun
#ikinci dereceden bir polinomun katsayılarını hesaplar.
#İkinci dereceden bir polinom genel olarak ax^2 + bx + c şeklinde ifade edilir.
#En uygun ikinci dereceden polinomun a, b ve c katsayılarını hesaplar.
katsayilar = np.polyfit(x_ust, y, 5)

# Polinomun y değerlerini hesapla
tahmin_y = np.polyval(katsayilar, x_ust)

print("Polinom katsayıları:", katsayilar)
print("Tahmin edilen y değerleri:", tahmin_y)
```

```python
import numpy as np
import matplotlib.pyplot as plt

#Örnek zaman serisi verisi
t = np.arange(1, 101)  # 1'den 100'e kadar olan zaman noktaları

#Basit bir doğrusal trend ile rastgele gürültü eklenmiş veri
#y = 3 * t + np.random.normal(0, 10, 100)
y = 3 * t + np.random.normal(0, 30, 100)

#Veriyi görselleştirme
plt.figure(figsize=(10, 6))
plt.scatter(t, y, label='Veri Noktaları', color='blue')
plt.xlabel('Zaman')
plt.ylabel('Değer')
plt.title('Trend Analizi: Örnek Zaman Serisi')
plt.legend()


#Doğrusal eğilim eğrisini hesaplama
coefficients = np.polyfit(t, y, 1) #ax+b
trend_line = np.polyval(coefficients, t) #y=ax+b

#Doğrusal eğilim eğrisini çizme
plt.plot(t, trend_line, label='Doğrusal Trend', color='red')
plt.legend()

plt.show()

expected_y = np.polyval(coefficients, [100,150,200])
#expected_y = np.polyval(coefficients, 200)
```
