Buna göre modele giriş olarak vermek üzere **X (features)** ve **y (label)** değişkenlerini tanımlayalım.

```python
y = data["Selling_Price"]

X = data.drop(["Selling_Price","Car_Name"], axis=1)
```

Eğitim ve test kümelerini %80-%20 olacak şekilde ayıralım ve eğitim ve test kümelerinin boyutlarını yazdıralım.

**train_test_split** fonksiyonu, veri setini eğitim ve test kümelerine böler. Bu fonksiyon, eğitim ve test kümelerinin boyutunu belirlemek için **test_size** parametresini kullanır. Ayrıca, verilerin bölünmesinin rastgele olmasını sağlamak için **random_state** parametresini kullanır.

Aşağıdaki kod satırı, X ve y veri setlerini eğitim ve test kümelerine böler. Eğitim kümesi, x_train ve y_train değişkenlerine atanır. Test kümesi ise, x_test ve y_test değişkenlerine atanır.

test_size parametresi, test kümesinin boyutunu belirtir. Bu parametre, 0 ile 1 arasında bir değer alabilir. Örneğin, test_size=0.2 değeri, test kümesinin boyutunun veri setinin %20'si olacağını belirtir.

random_state parametresi, verilerin bölünmesinin rastgele olmasını sağlar. Bu parametre, bir tamsayı değeri alabilir. Örneğin, random_state=1 değeri, verilerin bölünmesinin her zaman aynı şekilde yapılmasını sağlar. Bu, deneylerin tekrarlanabilirliğini artırır.

Eğitim ve test kümelerine bölmek, makine öğrenimi modellerinin performansını değerlendirmek için önemlidir. Eğitim kümesi, modeli eğitmek için kullanılır. Test kümesi ise, modelin performansını değerlendirmek için kullanılır.



```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(X, y, test_size=0.2, random_state=1)

print("x train: ", x_train.shape)
print("x test: ", x_test.shape)
print("y train: ", y_train.shape)
print("y test: ", y_test.shape)
```

Varsayılan (default) hiperparametreler ile LinearRegression modelini tanımlayalım ve eğitelim.

Doğrusal regresyon, bağımlı değişken ile bir veya daha fazla bağımsız değişken arasındaki doğrusal ilişkiyi modellemek için kullanılan bir makine öğrenimi algoritmasıdır.

x_train ve y_train değişkenleri, eğitim verilerini içerir. Eğitim verileri, modelin öğrenmesi için kullanılan verilerdir.

lr değişkeni, doğrusal regresyon modeli nesnesini içerir. Bu nesne, eğitim verileriyle eğitildikten sonra yeni veriler üzerinde tahminler yapmak için kullanılabilir.

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()

lr.fit(x_train, y_train)
```

Test veri kümesi ile tahmin sonuçlarını elde ediniz.

```python
x_test.iloc[0]
```

```python
predicts=lr.predict(x_test)
predicts
```

Orjinal ve prediction (tahmin) sonuçlarını aynı grafik üzerinde karşılaştırabilcek şekilde çizdirelim.

```python
y_test
```

```python
test_index=y_test.reset_index()["Selling_Price"]
test_index
```

```python
# Tahmin edilen değerler ile gerçek değerleri karşılaştıran görsel bir grafik

# Test verilerinin indeksi ve hedef değişkeni alalım
test_index=y_test.reset_index()["Selling_Price"]


# Gerçek değerleri kırmızı renkte ve 2 piksel kalınlığında çizgi olarak çizelim
ax=test_index.plot(label="originals",figsize=(12,6),linewidth=2,color="r")

# Tahmin edilen değerleri içeren bir DataFrame oluşturalım ve bu değerleri yeşil renkte ve 2 piksel kalınlığında çizgi olarak çizelim.
prediction=pd.DataFrame(predicts)
ax=prediction[0].plot(label = "predictions",figsize=(12,6),linewidth=2,color="g")

plt.legend(loc='upper right')
plt.title("ORIGINALS VS PREDICTIONS")
plt.xlabel("index")
plt.ylabel("values")

plt.show()
```
