**trend Analiziz & Satış Tahmini**

## Bir perakende şirketi, son birkaç yıldaki aylık satış verilerini içeren bir zaman serisi verisine sahiptir. Şirket, satışların genel bir artış eğilimi olduğunu fark etmiştir, ancak veri setinde bazı aykırı değerler ve gürültüler bulunmaktadır. Şirket, bu veri setindeki lineer bir trendi belirlemek ve gelecekteki satışları tahmin etmek istiyor.

## Bu senaryoda, Python kullanarak bu sorunu nasıl çözümleyeceğinizi açıklayınız. Adım adım düşünerek algortimayı tasarlayınız.

**Kullanabileceğiniz yapılar;**

polyfit ve polyval, Python'un NumPy kütüphanesi içinde bulunan çoklu dereceden polinom regresyonu yapmak için kullanılan fonksiyonlardır.

**polyfit:** Bu fonksiyon, bir veri setindeki x ve y değerlerine en uygun polinom katsayılarını hesaplar. Genellikle numpy.polyfit(x, y, derece) şeklinde kullanılır. x ve y, veri setindeki bağımsız ve bağımlı değişkenleri temsil eder. derece, kullanılacak polinomun derecesini belirtir.

**polyval:** Bu fonksiyon, verilen x değerleri için polinomun y değerlerini hesaplar. numpy.polyval(katsayılar, x) şeklinde kullanılır. katsayılar, polyfit tarafından hesaplanan polinom katsayılarını, x ise tahmin edilecek x değerlerini temsil eder.

```python
#Bu örnek, bir veri setindeki x, y değerlerine en uygun 2. dereceden
#bir polinomun katsayılarını hesaplar ve bu polinomun y değerlerini
#verilen x değerleri için tahmin eder.

import numpy as np

# Veri seti
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([0, 0.8, 0.9, 0.1, -0.8, -1])

# Polinom katsayılarını hesapla (2. dereceden polinom)
katsayilar = np.polyfit(x, y, 2)

# Polinomun y değerlerini hesapla
tahmin_y = np.polyval(katsayilar, x)

print("Polinom katsayıları:", katsayilar)
print("Tahmin edilen y değerleri:", tahmin_y)
```
